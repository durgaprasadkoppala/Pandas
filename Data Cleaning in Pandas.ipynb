```python
df = df.drop_duplicates()

df
```

```python
df = df.drop(columns = 'Not_Useful_Column')

df
```

```python
df["Last_Name"].str.lstrip("...")
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

```python
df
```

```python
 df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')
df
```

```python
#Changing the entire Phone_Number column float & str values to string 

for i in df['Phone_Number']:
    i = str(i)
    print(i , type(i))
```

```python
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')

df
```

```python
df[["Street Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand = True)

df
```

```python
df["Paying Customer"] = df["Paying Customer"].replace("N","No")
df["Paying Customer"] = df["Paying Customer"].replace("Y","Yes")

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Noo","No")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yeses","Yes")

df
```

```python
#Replacing N/a with " "

df = df.replace('N/a','')

df
```

```python
#filling all the null, NaN values with " "

df = df.fillna('')
df
```

```python
#Droping rows which are having 'Yes' in the Do_Not_Contact Column

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Yes':
        df.drop(x, inplace = True)
        
df
```

```python
#Droping rows which are having no phone number in the Phone_Number Column

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
        
df

#Another way to drop null values 
#df = df.dropna(subset="Phone_Number", inplace = True)
```

```python
# Reseting the Index values

df = df.reset_index(drop=True)
df
```
